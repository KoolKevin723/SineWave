# Sine Wave Loading

Edit the code below and hit shift + enter while in the code block to make the plot.

Remember:

y = A  sin(f  2  pi  x) + m
where:

* A is the amplitude
* f is the frequency
* m is the mean

Don't forget to check your units and change the labels if required.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,1,500)

A = 27.1874
f = 2
m = 0

F = A * np.sin(f * 2 * np.pi * x) + m

fig, ax1 = plt.subplots()
ax1.set_xlabel('time (s)')
ax1.set_ylabel('Load [kN]')
ax1.set_ylim(-40,40)
ax1.plot(x, F, 'b')
fig.tight_layout()
plt.show()
```
